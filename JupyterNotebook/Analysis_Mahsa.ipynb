```python
from datetime import datetime
import pandas as pd
```

```python
data = pd.read_csv('./Resources/data.csv', index_col='Unnamed: 0')
```

```python
data.head(2)
```

```python
data.info()
```

```python
data["timestamp"] = pd.to_datetime(data["timestamp"])
```

```python
data["date"] = data["timestamp"].map(lambda x: x.date())
```

```python
data["time"] = data["timestamp"].map(lambda x: x.time())
```

```python
data.head(2)
```

```python
target = data.groupby(["date"])["date"].count()
```

```python
#target = pd.DataFrame(data=target, index=None).reset_index()
target = target.loc[lambda x: x>10]
```

```python
dates = target.index.tolist()
```

```python
data.loc[data.date == dates[0]]
```

```python
def sametime(date):
    x = data.loc[data.date == date]
    n = x["country"].nunique()
    return n > 2
```

```python
valid_dates = []
for d in dates:
    if sametime(d) == True: 
        valid_dates.append(d)
```

```python
valid_dates 
```

```python
observ_one = data.loc[data.date == valid_dates[0]].sort_values(by="time")
```

```python
observ_one
```

```python
observ_one['shape'].unique()
```

```python
one = observ_one.to_json(orient="records")
with open("./Resources/dateOne.json", "w") as outfile:
    outfile.write(one)
```
